```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Quadratic Function

# Coefficients
a = 1
b = 3
c = 2

# Generate x values
x = np.linspace(0, 10, 100)

# Calculate y values
y = a * x**2 + b * x + c

# Create the plot
plt.plot(x, y, label="$y = x^2 + 3x + 2$")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Quadratic Function Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cubic Fuction

# Coefficients
a = 1
b = -6
c = 11
d = -6

# Generate x values
x = np.linspace(0, 10, 100)

# Calculate y values
y = a * x**3 + b * x**2 + c * x + d

# Create the plot
plt.plot(x, y, label="$y = x^3 - 6x^2 + 11x - 6$")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Cubic Function Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exponential Function

# Generate x values
x = np.linspace(0, 10, 100)

# Calculate y values
y = np.exp(x)

# Create the plot
plt.plot(x, y, label="$y = e^x$")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Exponential Function Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Logarithmic Function

# Generate x values
x = np.linspace(1, 10, 100)

# Calculate y values
y = np.log(x)

# Create the plot
plt.plot(x, y, label="$y = log(x)$")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Logarithmic Function Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def custom_sequence(start, stop, num, repidx: int = 0):
    """Renturn non-linear increasing sequence values

    Non-linear increasing plots
    0. cumsum - flex increasing
    1. exponential - late increasing
    2. logarithmic - early increasing
    3. quadratic - smooth increasing, coefficient
    4. cubic - smooth increasing, coefficient
    """

    kind = 5
    intervals = np.linspace(0.1, 1.1, num)

    match repidx % kind:
        case 0:  # cumsum
            np.random.shuffle(intervals)
            sequence = np.cumsum(intervals)

        case 1:  # exponential
            sequence = np.exp(intervals)
            np.random.shuffle(intervals)
            sequence = sequence + intervals

        case 2:  # logarithimic
            sequence = np.log(intervals)
            np.random.shuffle(intervals)
            sequence = sequence + intervals

        case 3:  # quadratic, slop depend on coefficent (a,b,c)
            a, b, c = 1, 3, 2
            sequence = a * intervals**2 + b * intervals + c
            np.random.shuffle(intervals)
            sequence = sequence + intervals

        case _:  # cubic
            a, b, c, d = 2, -6, 11, -6
            sequence = a * intervals**3 + b * intervals**2 + c * intervals + d
            np.random.shuffle(intervals)
            sequence = sequence + intervals

    sequence = start + (stop - start) * (sequence - sequence.min()) / (sequence.max() - sequence.min())
    return sequence
```

```python
fig = plt.figure()
fig.set_size_inches(12, 8)

ya = [custom_sequence(0.1, 1.1, 50, i) for i in range(5)]
for i in range(5):
    plt.plot(ya[i], label=f"repidx={i}")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Various Function Plot")

plt.legend()
plt.grid(True)
plt.show()
```
